```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from dtreeviz.trees import dtreeviz

data = pd.read_csv("https://raw.githubusercontent.com/e4rll/FUseminer_20211228/master/DT_sampleData.csv")
```

### 3.欠損値処理（failure_flgが空白の行を削除）

```python
data.dropna(subset = ["failure_flg"], inplace=True)
```

```python
data.describe()
```

### 3.ダミー変数変換

```python
data["place_product"].unique()
```

```python
data["place_purchase"].unique()
```

```python
data['place_product'].replace(to_replace=['Yokohama Factory', 
                                    'Saitama Factory', 'Mie Factory'], value=[1, 2, 3], inplace=True)
data['place_purchase'].replace(to_replace=['Hokkaido Region', 'Kanto Region', 'Kyushu Region', 
                                    'Shikoku Region', 'Kinki Region', 'Tohoku Region', 'Chubu Region'],
                                    value=[1, 2, 3, 4, 5, 6, 7], inplace=True)
```

```python
data.head()
```

### 4.目的変数・説明変数定義、決定木構築・可視化

```python
X = data.drop(columns=['failure_flg', "product_id"])
Y = data["failure_flg"]
```

```python
model = DecisionTreeClassifier(max_depth=3, random_state=0)
model.fit(X, Y)
# model.predict(X)
# sum(model.predict(X)==Y)/len(Y)  予測値・精度については今回は非表示→興味がある方は調べてみてください！
```

```python
viz = dtreeviz(model, X, Y,
               target_name = "failure_flg", 
               feature_names = ["place_product", "place_purchase", "avg_usetime_day_h"],
               class_names = ["Non-Trouble", "Trouble"],
               fontname = "DejaVu Sans")
```
